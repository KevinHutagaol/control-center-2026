```python
os = 19
ts = 0.1

zeta = -np.log(os/100) / np.sqrt(np.pi**2 + (np.log(os/100))**2)
wn = 4 / (zeta * ts)

print("Damping Ratio (zeta):", zeta)
print("Natural Frequency (wn):", wn)
```

```python
real = -zeta * wn
imaginary = wn * np.sqrt(1 - zeta**2)
pole_1 = complex(real, imaginary)

pole_2 = complex(real, -imaginary)

print("Desired Poles:", pole_1, pole_2)
```
